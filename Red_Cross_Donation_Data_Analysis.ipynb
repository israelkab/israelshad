```python
# Qualitiative Cleaning 
# Selecting only valid variables ignoring noise, redundant & constant variables

cols = ['ZIPCODE', 'AGE', 'MARITAL_STATUS', 'GENDER','ALUMNUS_IND', 'PARENT_IND', 'HAS_INVOLVEMENT_IND', 'WEALTH_RATING',
       'DEGREE_LEVEL', 'PREF_ADDRESS_TYPE', 'EMAIL_PRESENT_IND', 'Donor_years','DONOR_IND']
```

```python
# Setting the dataframe to contain only valid columns
df1 = df[cols]
```

```python
# To check number of nulls/missing data in the dataset
df1.isnull()
df1.isnull().sum()
```

```python
#Quantitative cleaning - 1
#dropping columns with missing more than 20% missing values (i.e. 6902)
df2 = df1.drop(columns = ['MARITAL_STATUS','WEALTH_RATING','DEGREE_LEVEL'])

df2.columns
```

```python
#Quantitative cleaning - 2
# Replacing specific missing values with mode in GENDER and PREF_ADDRESS_TYPE  columns

for column in ['GENDER', 'PREF_ADDRESS_TYPE']:
  mode = df2[column].mode()
  df2[column].fillna((mode)[0], inplace=True)
```
