```
#Qualitative Cleaning (Ignore noise variables)
cols = ['age','workclass','education-num','marital-status','occupation','relationship','race','sex','capital-gain','capital-loss','hours-per-week','native-country','income']
```

```
#Display valid columns in df1
df1 = df[cols]
df1.columns
```

```
#Quantative cleaning
df1.isnull().sum()
```

```
# To replace specific columns with Mode
for column in ['workclass', 'occupation', 'native-country']:
    #a = df1['workclass'].mode()[0]
    df1[column].fillna(df1[column].mode()[0], inplace=True)
```

```
# Display number of missing values by feature after all quantitative and qualititatve cleaning
df1.isnull().sum()
```

```
#Define categorical columns
cat_cols = ['workclass',	'marital-status',	'occupation',	'relationship',	'race',	'sex',	'native-country']
df2 = df1[cat_cols]   #df2 is my cat. column
df2.columns
```

```
# method 2 - Doing Auto ML with PyCaret
!pip install pycaret --upgrade
```

```
#To import all Classification related Pycaret libraries
from pycaret.classification import *
```

```
# Setting up the classification ML experiment
classification_model_setup = setup(df1,  target = 'income', 
                               categorical_features= cat_cols,
                               train_size = 0.7)
```

```
# Running the multiple algorithms
compare_models()
```

```
# Creating the model with the best algorithm based on the above results
model = create_model('lightgbm')
```

```
# Loading the Client dataset for which predictions to be made
pred_data = pd.read_csv('/content/Tarrant Dallas Denton Counties data.csv')
```

```
# Predicting the values using the best model
predictions = predict_model(model, data = pred_data)
predictions
```

```
# Converting predictions dataframe to csv
predictions.to_csv("Tarrant Dallas Denton Counties Income Predictions.csv")
```
